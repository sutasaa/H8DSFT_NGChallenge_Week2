Membuat koneksi antara pythoin dengan SQL Database dan diberikan nama database tersebut 'h8dsft_SQL.db'

```python
import sqlite3
conn = sqlite3.connect('h8dsft_SQL.db')
c = conn.cursor()
```

Membuat table 'assessment' dari data assessment.csv yang sudah di download dan menyimpannya di database

```python
import pandas as pd
users = pd.read_csv('assessments.csv')
# write the data to a sqlite table
users.to_sql('assessments', conn, if_exists='append', index = False)
```

Membuat query untuk memperoleh data dengan kolom: id_student, region, code_module, assessment_type, score. Data yang diambil hanya dengan code_presentation 2014B

```python
DataSQL = pd.read_sql('''SELECT sa.id_student, si.region, asm.code_module, asm.assessment_type, sa.score
                FROM studentassessment sa
                LEFT JOIN studentinfo si ON si.id_student = sa.id_student
                LEFT JOIN assessments asm ON asm.id_assessment = sa.id_assessment
                WHERE asm.code_presentation = '2014B'
                ''', conn)
DataSQL
```

Menyimpan data frame ke dalam excel

```python
nama_file = "NGChallenge3.xlsx"
DataSQL.to_excel(nama_file)
print("DataFrame ke Excel telah berhasil")
```
